# Descriptive Statistics and Plots

```python
%matplotlib inline
combined_data = combined_data.dropna(subset= ['Year'])
combined_data['Year'] = combined_data['Year'].astype(int)
fatal_accidents_by_year = combined_data.groupby(['Year', 'FatalIndicator']).size().unstack().plot(kind='bar', stacked=True, figsize = (12,6), title='Fatal vs. Non-Fatal Accidents by Year')
```

```python
fatal_accidents_by_month = combined_data.groupby(['Month', 'FatalIndicator']).size().unstack().plot(kind='bar', stacked=True, figsize = (12,6), title='Fatal vs. Non-Fatal Accidents by Month of Year')
```

```python
#Monday is 0 - Sunday is 6
fatal_accidents_by_weekday = combined_data.groupby(['WeekDay', 'FatalIndicator']).size().unstack().plot(kind='bar', stacked=True, figsize = (12,6), title='Fatal vs. Non-Fatal Accidents by Day of Week')
```

```python
combined_data['AircraftCategory'] = combined_data['AircraftCategory'].fillna('missing')
aircraft_category_by_year = combined_data.groupby(['Year', 'AircraftCategory']).size().unstack().plot(kind='bar', stacked=True,figsize = (12,6), title='Aircraft Category by Year')
```

```python
combined_data['WeatherCondition'] = combined_data['WeatherCondition'].fillna('missing')
weather_condition_fatality = combined_data.groupby(['WeatherCondition', 'FatalIndicator']).size().unstack().plot(kind='bar', stacked=True,figsize = (12,6), title='Fatality by Weather Condition')
```

```python
combined_data['Make'] = combined_data['Make'].fillna('missing')
combined_data['Make'] = combined_data['Make'].str.lower()
make_accident_freq = combined_data['Make'].value_counts()[:20].plot(kind='barh',figsize = (12,6), title='20 most common aircraft makes')
```

```python
combined_data['Model'] = combined_data['Model'].fillna('missing')
combined_data['Model'] = combined_data['Model'].str.lower()
model_accident_freq = combined_data['Model'].value_counts()[:20].plot(kind='barh',figsize = (12,6), title='20 most common aircraft models')
```

```python
combined_data["MakeModel"] = combined_data["Make"] + " " + combined_data["Model"]
makemodel_accident_freq = combined_data['MakeModel'].value_counts()[:40].plot(kind='barh',figsize = (12,6), title='40 most common aircraft make/models')
```

```python
combined_data_fatal = combined_data[combined_data['FatalIndicator'] == 1]
makemodel_fatal_accident_freq = combined_data_fatal['MakeModel'].value_counts()[:40].plot(kind='barh',figsize = (12,6), title='40 most common aircraft make/models')
```

```python
combined_data['Country'] = combined_data['Country'].fillna('missing')
combined_data['Country'].value_counts().plot(kind='pie',figsize = (8,8), autopct='%.2f', title = 'Accidents by Country')
```

```python
combined_data['ForeignIndicator'] = combined_data['Country'].apply(lambda x: 1 if x != 'United States' else 0)
```

## US Map of Unscaled Data

```python
fig = plt.figure(figsize=(12, 12))
m = Basemap(llcrnrlon=-121,llcrnrlat=20,urcrnrlon=-62,urcrnrlat=51,
    projection='lcc',lat_1=32,lat_2=45,lon_0=-95)

# data from U.S Census Bureau
shp_info = m.readshapefile('Documents/Spring_2020/data-scientist-exercise02/st99_d00','states',drawbounds=True)

# choose a color for each state based on population density.
colors={}
statenames=[]
cmap = plt.cm.get_cmap('Reds_r') # use 'hot' colormap
vmin = 0; vmax = 10000 # set range.
for shapedict in m.states_info:
    statename = shapedict['NAME']
    # skip DC and Puerto Rico.
    if statename not in ['District of Columbia','Puerto Rico']:
        short_name_index = list(short_state_names.values()).index(statename)
        short_name = list(short_state_names.keys())[short_name_index]
        accidents = US_state_count_dict[short_name]
        # calling colormap with value between 0 and 1 returns
        # rgba value.  Invert color range (hot colors are high
        # population), take sqrt root to spread out colors more.
        colors[statename] = cmap(1.-np.sqrt((accidents-vmin)/(vmax-vmin)))[:3]
    statenames.append(statename)
# cycle through state names, color each one.
ax = plt.gca() # get current axes instance

for nshape,seg in enumerate(m.states):
    # skip DC and Puerto Rico.
    if statenames[nshape] not in ['District of Columbia','Puerto Rico']:
        # Offset Alaska and Hawaii to the lower-left corner. 
        if statenames[nshape] == 'Alaska':
        # Alaska is too big. Scale it down to 35% first, then transate it. 
            seg = list(map(lambda x: (0.35*x[0] + 1100000, 0.35*x[1]-1350000), seg))
        if statenames[nshape] == 'Hawaii':
            seg = list(map(lambda x: (x[0] + 5100000, x[1]-1450000), seg))
        color = rgb2hex(colors[statenames[nshape]]) 
        poly = Polygon(seg,facecolor=color,edgecolor=color)
        ax.add_patch(poly)

plt.title('Filling State Polygons by Accident Density')
plt.show()
```

## Clustering Accidents

```python
X = combined_data[['WeatherCondition','AmateurBuilt','BroadPhaseOfFlight','NumberOfEngines', 'PurposeOfFlight']]
```

```python
X= X.fillna('missing')
```

```python
X['NumberOfEngines'] = X['NumberOfEngines'].astype(str)
```

```python
#label encode data before clustering
cat_features = ['WeatherCondition','AmateurBuilt','BroadPhaseOfFlight','NumberOfEngines', 'PurposeOfFlight']
encoder = LabelEncoder()
X = X[cat_features].apply(encoder.fit_transform)
```

```python
clusters = 4
kmeans = KMeans(n_clusters= clusters, init='k-means++', max_iter=300, n_init=10, random_state=0)
kmeans.fit(X)
```

```python
#use pca to better visualize clusters
pca = PCA(n_components = 2)
axes = pd.DataFrame(pca.fit_transform(X))
```

```python
X['Labels'] = kmeans.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(axes.iloc[:,0], axes.iloc[:,1], hue=X['Labels'], 
                palette=sns.color_palette('hls', 4))
plt.title('KMeans with 4 Clusters on Structured Data')
plt.show()
```
